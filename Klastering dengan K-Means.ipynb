```python
#Import Library Python yang diperlukan
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as numpy
import pandas as pd
from sklearn.cluster import KMeans
```

```python
#Import Dataset
dataset = pd.read_csv("dataset_kmeans.csv")
dataset.keys()
```

```python
data = pd.DataFrame(dataset)
data.head()
```

```python
#Konversi ke Data Array
X = np.asarray(dataset)
print(X)
```

```python
#Menampilkan Data Dalam Bentuk Scatter Plot
plt.scatter(X[:,0], X[:,1], label='True Position')
plt.xlabel("INCOME")
plt.ylabel("SPEND")
plt.title("Grafik Penyebaran Data")
plt.show()
```

```python
#Mengaktifkan K-Means dengan Jumlah K=2
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
#Menampilkan Nilai Centroid yang Digenerate oleh Algoritma
print(kmeans.cluster_centers_)
```

```python
#Plot data Point
#Memvisualisasikan Hasil Klasterisasi Data
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_, cmap='rainbow')
plt.xlabel("INCOME")
plt.ylabel("SPEND")
plt.title("Grafik Hasil Klasterisasi Data Income dan Spend")
plt.show()
```

```python
#Plot Data Point
#Memvisualisasikan Hasil Klasterisasi dengan Centroid dari masing-masing Klaster
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color='black')
plt.xlabel("INCOME")
plt.ylabel("SPEND")
plt.title("Grafik Hasil Klasterisasi Data Income dan Spend")
plt.show()
```
