```python
# Import Library
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
#Import Dataset
data = pd.read_csv('dataset_AHC.csv')
```

```python
data.head()
```

```python
x = np.array(data)
print(x)
```

```python
#Melakukan hierachical klastering
Z = linkage(x)
```

```python
#Plot dendrogram
plt.figure(figsize=(30, 10))
plt.title('Dendrogram Hierarchical Klastering')
plt.xlabel('data')
plt.ylabel('jarak')
dendrogram(Z)
plt.show()
```
