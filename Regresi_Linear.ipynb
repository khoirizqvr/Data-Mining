```python
# Import Library
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
from scipy import stats
```

```python
#Import Dataset
data = pd.read_csv('dataset_regresi_linear.csv')
x = data.iloc[:, :-1].values
y = data.iloc[:, 1].values
```

```python
data.keys()
```

```python
data.shape
```

```python
#Menampilkan Isi Sebagian Dataset
dt = pd.DataFrame(data)
dt.head()
```

```python
# Split Dataset Menjadi Training Set & Testing Set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=0)
```

```python
#Melakukan Fitting Simple Linear Regression pada Training Set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
#Memprediksi Hasil Test-Set
y_pred = regressor.predict(x_test)
```

```python
#Visualisasi Data
plt.scatter(dt.Experinces, dt.Salary)
plt.xlabel("Experinces")
plt.ylabel("Salary")
plt.title("Grafik Experinces vs Salary")
plt.show()
```

```python
#Visualisasi Hasil Prediksi Pada Training-Set
#Ukuran PLot
plt.figure(figsize=(10,8))

#Kuning adalah data observasi
plt.scatter(x_train, y_train, color='gre')

#Garis Biru adalah hasil prediksi dari machine learning
plt.plot(x_train, regressor.predict(x_train), color='blue')

#Memberi judul & label
plt.title('Training Set')
plt.xlabel('Experinces')
plt.ylabel('Salary')

#Menampilkan plot
plt.show()
```
